```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def signalCreate(A, w, d, t):
    return A*np.sin(w*2.0*np.pi*t + d)

def FFTAnalysis(signal):
    transformed = np.fft.fft(signal)
    
    transformed = 2.0/(len(signal))* np.abs(transformed[:len(signal)/2])
    
    return transformed
    
def frequencyRange(signalTime):
    deltaT = []
    for i in range(len(signalTime)-1):
        deltaT.append(signalTime[i+1]-signalTime[i])
    avgDeltaT = np.mean(deltaT)
    
    freq = np.linspace(0.0, 1.0/(2.0*avgDeltaT), len(signalTime)/2)
    
    return freq
```

```python
time = np.linspace(0.0, 60.0, 1000)
testSignal = signalCreate(5, 0.5, 0, time)


frequency = frequencyRange(time)

transTestSignal = FFTAnalysis(testSignal)

plt.figure(figsize=(10,8))
plt.plot(time, testSignal)
plt.grid()
plt.show()

plt.figure(figsize=(10,8))
plt.plot(frequency, transTestSignal)
plt.grid()
plt.show()
```

```python
time = np.linspace(0.0, 30.0, 1000)
testSignal = signalCreate(5, 2.0, 0, time)


frequency = frequencyRange(time)

deltaT = []
for i in range(len(time)-1):
    deltaT.append(time[i+1]-time[i])
avgDeltaT = np.mean(deltaT)
    
frequencyLong = np.linspace(0.0, 1.0/(2.0*avgDeltaT), len(time))



transformed = np.fft.fft(testSignal)
    
transformed = 2.0/(len(testSignal))* np.abs(transformed[:len(testSignal)/2])

transformedBase = np.fft.fft(testSignal)

transformedCropped = np.fft.fft(testSignal)
transformedCropped = transformedCropped[:len(testSignal)/2]

transformedCroppedAbs = np.fft.fft(testSignal)
transformedCroppedAbs = np.abs(transformed[:len(testSignal)/2])

plt.figure(figsize=(10,8))
plt.title("Original Signal")
plt.plot(time, testSignal)
plt.grid()
plt.show()

plt.figure(figsize=(10,8))
plt.title("FFT Signal")
plt.plot(frequency, transformed)
plt.grid()
plt.show()

plt.figure(figsize=(10,8))
plt.title("Transformed Signal, No Crop or Scale")
plt.plot(frequencyLong, transformedBase)
plt.grid()
plt.show()

plt.figure(figsize=(10,8))
plt.title("Transformed Cropped Signal, No Scale")
plt.plot(frequency, transformedCropped)
plt.grid()
plt.show()

plt.figure(figsize=(10,8))
plt.title("Absolute Value Transformed Cropped Signal, No Scale")
plt.plot(frequency, transformedCroppedAbs)
plt.grid()
plt.show()
```
