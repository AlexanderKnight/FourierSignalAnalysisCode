```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

data = np.genfromtxt('2014_02_24-run01Data.txt', delimiter='\t')

time = data[:,0]
sumSignal = data[:,1]
LR_Signal = data[:,2]
TB_Signal = data[:,3]

sumSignalF = np.fft.fft(sumSignal)
LR_SignalF = np.fft.fft(LR_Signal)
TB_SignalF = np.fft.fft(TB_Signal)

deltaT = []
for i in range(len(time)-1):
    deltaT.append(time[i+1]-time[i])

avgDeltaT = np.mean(deltaT)

plt.plot(time,sumSignal)
plt.show()

plt.plot(time,LR_Signal)
plt.show()

plt.plot(time,TB_Signal)
plt.show()

timeF = np.linspace(0.0, 1.0/(2.0*avgDeltaT),len(time)/2)

plt.plot(timeF,np.abs(sumSignalF[:len(time)/2])/max(sumSignalF))
plt.grid()
plt.ylim(0,0.02)
plt.xlim(0,0.2)
plt.show()

plt.plot(timeF,np.abs(LR_SignalF[:len(time)/2])/max(LR_SignalF))
plt.grid()
plt.ylim(0,0.02)
plt.show()

plt.plot(timeF,np.abs(TB_SignalF[:len(time)/2])/max(TB_SignalF))
plt.grid()
plt.ylim(0,0.05)
plt.xlim(0,8)
plt.show()
```
