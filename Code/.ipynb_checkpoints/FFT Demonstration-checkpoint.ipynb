# FFT Demonstration

## Fourier Transforms

A Fourier Transform is a mathematical method to take a time dependant signal and break it down into a sum of sinusoidal frequences. The tranformed function, denoted $\hat{\mathscr{f}}$ is defined:

$$\hat{\mathscr{f}}(\omega) = \int^{\infty}_{-\infty}\mathscr{f}(x)e^{-2\pi i x \omega} dx$$

The inverse transformation function is:

$$\mathscr{f}(x) = \int^{\infty}_{-\infty} \hat{\mathscr{f}}(\omega)e^{2\pi i \omega x} d \omega $$

There is a limit on the maximum frequency that can be detected from this. This limit is dependant on the average time sample rate.

It is :

$$\omega_{max} = \frac{1}{2 \bar{\Delta t}} $$

Where $\bar{\Delta t}$ is your average sampling time.



```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def signalCreate(A, w, d, t):
    return A*np.sin(w*2.0*np.pi*t + d)

def FFTAnalysis(signal):
    transformed = np.fft.fft(signal)
    
    transformed = 2.0/(len(signal))* np.abs(transformed[:len(signal)/2])
    
    return transformed
    
def frequencyRange(signalTime):
    deltaT = []
    for i in range(len(signalTime)-1):
        deltaT.append(signalTime[i+1]-signalTime[i])
    avgDeltaT = np.mean(deltaT)
    
    freq = np.linspace(0.0, 1.0/(2.0*avgDeltaT), len(signalTime)/2)
    
    return freq
```

Generally, programs to analysis a signal should look like so:

```python
time = np.linspace(0.0, 60.0, 1000)


testSignal = signalCreate(5, 0.5, 0, time)


frequency = frequencyRange(time)

transTestSignal = FFTAnalysis(testSignal)

plt.figure(figsize=(10,8))
plt.plot(time, testSignal)
plt.grid()
plt.show()

plt.figure(figsize=(10,8))
plt.plot(frequency, transTestSignal)
plt.grid()
plt.show()

```

## Time Sampling

Time sampling is important, as np.fft.fft assumes that time is sampled evenly. If this is not the case, you get a much less precise graph, as shown below.

```python
diffStepTime = [0.0]
while max(diffStepTime)<60.0:
    diffStepTime.append(np.random.random()*0.1+max(diffStepTime))
    
diffTestSignal = signalCreate(5, 0.5, 0, np.array(diffStepTime))

stepFreq = frequencyRange(diffStepTime)

stepTransTestSignal = FFTAnalysis(diffTestSignal)

plt.figure(figsize=(10,8))
plt.plot(diffStepTime, diffTestSignal)
plt.grid()
plt.show()

plt.figure(figsize=(10,8))
plt.plot(stepFreq, stepTransTestSignal)
plt.grid()
plt.show()
```

```python
time = np.linspace(0.0, 60.0, 10000)
testSignal = signalCreate(5, 20, 0, time) + signalCreate(3, 40, 0, time) + signalCreate(2, 60, 0 , time)


frequency = frequencyRange(time)

transTestSignal = FFTAnalysis(testSignal)

plt.figure(figsize=(10,8))
plt.plot(time, testSignal)
plt.grid()
plt.show()

plt.figure(figsize=(10,8))
plt.plot(frequency, transTestSignal)
plt.grid()
plt.show()
```

```python
time = np.linspace(0.0, 30.0, 1000)
testSignal = signalCreate(5, 2.0, 0, time)


frequency = frequencyRange(time)

deltaT = []
for i in range(len(time)-1):
    deltaT.append(time[i+1]-time[i])
avgDeltaT = np.mean(deltaT)
    
frequencyLong = np.linspace(0.0, 1.0/(2.0*avgDeltaT), len(time))



transformed = np.fft.fft(testSignal)
    
transformed = 2.0/(len(testSignal))* np.abs(transformed[:len(testSignal)/2])

transformedBase = np.fft.fft(testSignal)

transformedCropped = np.fft.fft(testSignal)
transformedCropped = transformedCropped[:len(testSignal)/2]

transformedCroppedAbs = np.fft.fft(testSignal)
transformedCroppedAbs = np.abs(transformed[:len(testSignal)/2])

plt.figure(figsize=(10,8))
plt.title("Original Signal")
plt.plot(time, testSignal)
plt.grid()
plt.show()

plt.figure(figsize=(10,8))
plt.title("FFT Signal")
plt.plot(frequency, transformed)
plt.grid()
plt.show()

plt.figure(figsize=(10,8))
plt.title("Transformed Signal, No Crop or Scale")
plt.plot(frequencyLong, transformedBase)
plt.grid()
plt.show()

plt.figure(figsize=(10,8))
plt.title("Transformed Cropped Signal, No Scale")
plt.plot(frequency, transformedCropped)
plt.grid()
plt.show()

plt.figure(figsize=(10,8))
plt.title("Absolute Value Transformed Cropped Signal, No Scale")
plt.plot(frequency, transformedCroppedAbs)
plt.grid()
plt.show()
```

```python
time = np.linspace(0.0, 62.0, 1000)
testSignal = signalCreate(5, 0.1, 2, time)
normTestSignal = testSignal*np.sin(time/max(time)*np.pi)

freq = frequencyRange(time)
transSignal = FFTAnalysis(testSignal)
normTransSignal = FFTAnalysis(normTestSignal)

plt.plot(time, testSignal)
plt.show()

plt.plot(freq, transSignal)
plt.show()

plt.plot(time, normTestSignal)
plt.show()

plt.plot(freq, normTransSignal)
plt.show()
```
