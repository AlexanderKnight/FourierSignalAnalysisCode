```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

data = np.genfromtxt('xWobbleError.txt', delimiter=' ')

time = data[:,0]
xWobble = data[:,1]


xFWobble = np.fft.rfft(xWobble)
xFWobble = 2.0/len(xWobble) * np.abs(xFWobble)
freqx = np.fft.rfftfreq(time.shape[-1], d=max(time)/len(time))
```

```python
plt.plot(freqx,xFWobble)
plt.xlim(0.0, 8.0)
plt.grid()
plt.show()
```
